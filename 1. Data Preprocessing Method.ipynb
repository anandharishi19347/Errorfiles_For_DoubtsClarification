```python
#For One Hot Encoding Part (Split columns)

from IPython.display import display

# Load dataset
df = pd.read_csv("customer_churn_prediction_dataset.csv")

# Extract the two columns
df_payment_contract = df[['PaymentMethod', 'Contract']]

# Remove the two columns from the main dataset
df_remaining = df.drop(columns=['PaymentMethod', 'Contract'])
```

```python
df_payment_contract
```

```python
df_remaining
```

```python
# One-Hot Encoding with "No" & "Yes" format
df_payment_contract_encoded = df_payment_contract.copy()

for col in df_payment_contract.columns:
    encoded = pd.get_dummies(df_payment_contract[col], dtype=str)
    encoded = encoded.replace({"0": "No", "1": "Yes"})  # Replace 0/1 with "No"/"Yes"

    
    
    df_payment_contract_encoded = df_payment_contract_encoded.drop(columns=[col]).join(encoded)
```

```python
df_payment_contract_encoded
```

```python
#For replace No
df_payment_contract_encoded = df_payment_contract_encoded.replace("", "No")
```

```python
df_payment_contract_encoded
```

```python
# Merge the encoded payment & contract data with the remaining dataset
dataset = pd.concat([df_payment_contract_encoded,df_remaining], axis=1)
```

```python
dataset.head()
```

```python
#Remove Unnecessory columns
dataset = dataset.drop( ['customerID'],axis=1)
```

```python
dataset.head()
```

```python
dataset.replace({'No internet service': 'No', 'No phone service': 'No'}, inplace=True)
dataset.replace({'DSL': 'Yes', 'Fiber optic': 'Yes'}, inplace=True)
```

```python
dataset.head()
```

```python
# Split quan and qual values
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
       # print(columnName)
        if(dataset[columnName].dtype=='O'):
            #print("qual")
            qual.append(columnName)
        else:
           # print("quan")
            quan.append(columnName)
    return quan,qual
```

```python
# For check these all
descriptive=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99","Q4:100%","IQR",
                               "1.5Rule","Lesser","Greater","Min","Max"],columns=quan)
for coulmnName in quan:
    descriptive[coulmnName]["Mean"]=dataset[coulmnName].mean()
    descriptive[coulmnName]["Median"]=dataset[coulmnName].median()
    descriptive[coulmnName]["Mode"]=dataset[coulmnName].mode()[0]
    descriptive[coulmnName]["Q1:25%"]=dataset.describe()[coulmnName]["25%"]
    descriptive[coulmnName]["Q2:50%"]=dataset.describe()[coulmnName]["50%"]
    descriptive[coulmnName]["Q3:75%"]=dataset.describe()[coulmnName]["75%"]
    descriptive[coulmnName]["99"]=np.percentile(dataset[coulmnName],99)
    descriptive[coulmnName]["Q4:100%"]=dataset.describe()[coulmnName]["max"]
    descriptive[coulmnName]["IQR"]= descriptive[coulmnName]["Q3:75%"]-descriptive[coulmnName]["Q1:25%"]
    descriptive[coulmnName]["1.5Rule"]=1.5*descriptive[coulmnName]["IQR"]
    descriptive[coulmnName]["Lesser"]=descriptive[coulmnName]["Q1:25%"]-descriptive[coulmnName]["1.5Rule"]
    descriptive[coulmnName]["Greater"]=descriptive[coulmnName]["Q3:75%"]+ descriptive[coulmnName]["1.5Rule"]
    descriptive[coulmnName]["Min"]=dataset[coulmnName].min()
    descriptive[coulmnName]["Max"]=dataset[coulmnName].max()
```
