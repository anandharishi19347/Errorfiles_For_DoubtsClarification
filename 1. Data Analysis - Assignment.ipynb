# 7. Test the analysis of variance between etest_P and mba_p at significance level 5%. (make decision using Hypothesis Testing)

```python
dataset
```

```python
import scipy.stats as stats
from scipy.stats import f_oneway

dataset = pd.DataFrame(dataset)

# Perform one-way ANOVA
f_statistic, p_value = f_oneway(dataset['etest_p'], dataset['mba_p'])

# Decision based on p-value
alpha = 0.05
if p_value < alpha:
    print(f"Reject the null hypothesis and Accept the Alternate hypothesis (p-value = {p_value:.4f}). There is a significant difference.")
else:
    print(f"Fail to reject the null hypothesis (p-value = {p_value:.4f}). No significant difference.")
```

# Answers:

# Reject the null hypothesis and Accept the Alternate hypothesis (p-value = 0.0000). There is a significant difference.

```python
 #standard normal distribution
def stdNBgraph(dataset):
 # Coverted to standard Normal Distribution
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()

   
                
    values=[i for i in dataset]
                
    z_score=[((j-mean)/std) for j in values]
                
    sns.distplot(z_score,kde=True)
                
    sum(z_score)/len(z_score)
   # z_score.std()
```

# 10. What is the probability density function of the salary range from 700000 to 900000?

```python
dataset
```

```python
#find out probability density function value
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # generate a sample
    sample = dataset
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    #values = [value for value in range(startrange, endrange)]
    
    values = []
    for value in range(startrange, endrange):
        values.append(value)
    
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
get_pdf_probability(dataset["salary"],700000,900000)
```

# 13. plot any useful graph and explain it.

```python
dataset
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Convert the dataset to a DataFrame
dataset_df = pd.DataFrame(dataset, columns=['salary'])

# Ensure the column is numeric
dataset_df['salary'] = pd.to_numeric(dataset_df['salary'], errors='coerce')

# Drop any non-numeric values (if any)
dataset_df = dataset_df.dropna()

# Calculate mean, median, and mode
mean = dataset_df['salary'].mean()
median = dataset_df['salary'].median()
mode_series = dataset_df['salary'].mode()

# Handle multiple modes (if any)
mode = mode_series.iloc[0] if not mode_series.empty else None

# Plot the data
plt.hist(dataset_df['salary'], bins=8, alpha=0.7, color='blue', edgecolor='black')
plt.axvline(mean, color='red', label=f'Mean: {mean:.2f}')
plt.axvline(median, color='green', label=f'Median: {median:.2f}')

# Plot mode only if it exists
if mode is not None:
    plt.axvline(mode, color='orange', label=f'Mode: {mode:.2f}')

plt.legend()
plt.title('Mean, Median, and Mode in the Dataset')
plt.xlabel('salary')
plt.ylabel('Frequency')
plt.show()
```
